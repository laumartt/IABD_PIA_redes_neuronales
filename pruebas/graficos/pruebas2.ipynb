```python
import matplotlib.pyplot as plt
import numpy as np

figure=plt.figure(figsize=(8,4))
figure.suptitle("TÃ­tulo de Figure", fontsize=14, color='red')

'''subfigure_a,subfigure_b = figure.subfigures(nrows=2, ncols=1)
 
subfigure_a.suptitle("Titulo de la SubFigura A")'''

axes = figure.add_subplot(1,1,1)
x=np.array([0,1,2,3,4,5,6])
y=np.array([0,1,4,9,16,25,36])
axes.plot(x,y, label='datos1', color='red')

x=np.array([0,1,2,3,4,5,6])
y=np.array([0,9,7,9,0,34,8])
axes.plot(x,y, label='datos2', color='blue')
x=x[[1,5]]
y=y[[1,5]]
axes.scatter(x,y)
axes.set_title("TÃ­tulo de Figure", fontsize=11, color='blue')
axes.set_xlim(xmin=0, xmax=10)
axes.set_ylim(ymin=0, ymax=50)

#PERSONALIZACION
#Titulos
axes.set_xlabel('Mes/AÃ±o', fontsize=15,labelpad=20,color="#003B80")  
axes.set_ylabel('Importe (â¬)', fontsize=15,labelpad=20,color="#003B80")
axes.set_title("Gastos y Beneficios Mensuales", fontsize=20,pad=30,color="#003B80")
#Leyenda
axes.legend(fontsize=15,facecolor='#CDCDCD',labelcolor="#000000")
#Color fondo
axes.set_facecolor("#F0F7FF")
#Estilo ticks
axes.tick_params(axis='x',labelsize=13, colors="#FF00FF")
axes.tick_params(axis='y',labelsize=13, colors="#FF0000")
#CuadrÃ­cula
axes.grid(visible=True, which='major', axis='both',color="#A0A0A0",linewidth=1)
axes.set_axisbelow(True)
#Ejes
axes.spines['right'].set_visible(False)
axes.spines['top'].set_visible(False)
axes.spines['left'].set_color("#FF0000")
axes.spines['bottom'].set_color("#FF0000")
```
